Q.6: You can now segment users into distinct groups for targeted ad campaigns. Segmentation can be based on various criteria, such as:

Demographics:

Age, Gender, Income Level, Education Level

Behavioural:

Time Spent Online, Likes and Reactions, CTR, Conversion Rates

Interests:

Aligning ad content with the top interests identified
To implement user profiling and segmentation, apply clustering techniques or develop personas based on the combination of these attributes.

Start by selecting a subset of features that could be most indicative of user preferences and
behaviour for segmentation and apply a clustering algorithm to create user segments:

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import numpy as np

# Function to convert categorical age and income to numerical categories
def convert_categorical_to_numeric(df, column, mapping):
    df[column] = df[column].map(mapping)
    return df

# Age mapping
age_mapping = {
    '18-24': 1,
    '25-34': 2,
    '35-44': 3,
    '45-54': 4,
    '55-64': 5,
    '65+': 6
}
df = convert_categorical_to_numeric(df, 'Age', age_mapping)

# Income Level mapping
income_mapping = {
    '0-20k': 1,
    '20k-40k': 2,
    '40k-60k': 3,
    '60k-80k': 4,
    '80k-100k': 5,
    '100k+': 6
}
df = convert_categorical_to_numeric(df, 'Income Level', income_mapping)

# Selecting features for segmentation
features = df[['Age', 'Gender', 'Education Level', 'Income Level', 'Time Spent Online (hrs/weekday)',
               'Time Spent Online (hrs/weekend)', 'Likes and Reactions', 'Click-Through Rates (CTR)',
               'Conversion Rates', 'Ad Interaction Time (sec)', 'Top Interests']]

# Preprocessing: Encoding categorical features and scaling numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Age', 'Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)',
                                   'Likes and Reactions', 'Click-Through Rates (CTR)', 'Conversion Rates',
                                   'Ad Interaction Time (sec)', 'Income Level']),
        ('cat', OneHotEncoder(), ['Gender', 'Education Level', 'Top Interests'])
    ])

# Determine the optimal number of clusters using the Elbow method
scaled_data = preprocessor.fit_transform(features)
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()

# Fit the K-Means algorithm with the optimal number of clusters (let's assume k=5 for this example)
kmeans = KMeans(n_clusters=5, random_state=42)
cluster_labels = kmeans.fit_predict(scaled_data)
df['Cluster'] = cluster_labels

# Calculate mean and mode for clusters
cluster_summary = df.groupby('Cluster').agg({
    'Age': 'mean',
    'Time Spent Online (hrs/weekday)': 'mean',
    'Time Spent Online (hrs/weekend)': 'mean',
    'Likes and Reactions': 'mean',
    'Click-Through Rates (CTR)': 'mean',
    'Conversion Rates': 'mean',
    'Ad Interaction Time (sec)': 'mean',
    'Gender': lambda x: x.mode()[0],
    'Education Level': lambda x: x.mode()[0],
    'Income Level': 'mean',
    'Top Interests': lambda x: x.mode()[0]
}).reset_index()

print(cluster_summary)
```

Q.7: **Now compute the mean values of the numerical features and the mode for categorical features within each cluster to get a sense of their defining characteristics**

```python
import pandas as pd
import numpy as np
import plotly.express as px

# Prepare data for radar chart
# Normalize numerical data to fit the radar chart scale (0 to 1)
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

# Select numerical columns and normalize them
numerical_cols = ['Age', 'Time Spent Online (hrs/weekday)', 'Time Spent Online (hrs/weekend)',
                  'Likes and Reactions', 'Click-Through Rates (CTR)', 'Conversion Rates',
                  'Ad Interaction Time (sec)', 'Income Level']

cluster_numerical_means = cluster_summary[numerical_cols]
cluster_numerical_means_normalized = normalize(cluster_numerical_means)

# Add the cluster labels
cluster_numerical_means_normalized['Cluster'] = cluster_summary['Cluster']

# Melt the dataframe to fit the radar chart format
melted_cluster_data = cluster_numerical_means_normalized.melt(id_vars='Cluster', var_name='Feature', value_name='Value')

# Create radar chart
fig = px.line_polar(melted_cluster_data, r='Value', theta='Feature', color='Cluster', line_close=True,
                    title='Radar Chart of Cluster Characteristics')
fig.update_traces(fill='toself')
fig.show()
```

Q.8: **Now assign each cluster a name that reflects its most defining characteristics based on the mean values of numerical features and the most frequent categories for categorical
features.
Based on the cluster analysis, summarize and name the segments as follows:
Cluster 0  Weekend Warriors: High weekend online activity, moderate likes and
reactions, predominantly male, age group 25-34, income level 80k-100k.
1.
Cluster 1  Engaged Professionals: Balanced online activity, high likes and reactions,
predominantly male, age group 25-34, high income (100k+).
2.
Cluster 2  Low-Key Users: Moderate to high weekend online activity, moderate likes and
reactions, predominantly male, age group 25-34, income level 60k-80k, lower CTR.
3.
Cluster 3  Active Explorers: High overall online activity, lower likes and reactions,
predominantly female, age group 25-34, income level 60k-80k.
4.
Cluster 4  Budget Browsers: Moderate online activity, lowest likes and reactions,
predominantly female, age group 25-34, lowest income level (0-20k), lower CTR.**

```python
# Cluster Naming and Summary
def assign_cluster_names(cluster_summary):
    cluster_names = {}
    for index, row in cluster_summary.iterrows():
        if row['Time Spent Online (hrs/weekend)'] > row['Time Spent Online (hrs/weekday)']:
            if row['Likes and Reactions'] >= 50 and row['Income Level'] > 4:
                cluster_names[row['Cluster']] = 'Engaged Professionals'
            else:
                cluster_names[row['Cluster']] = 'Weekend Warriors'
        elif row['Likes and Reactions'] < 30 and row['Income Level'] < 3:
            cluster_names[row['Cluster']] = 'Budget Browsers'
        elif row['Click-Through Rates (CTR)'] < 0.1 and row['Income Level'] >= 3:
            cluster_names[row['Cluster']] = 'Low-Key Users'
        else:
            cluster_names[row['Cluster']] = 'Active Explorers'
    return cluster_names

# Assign names to clusters
cluster_names = assign_cluster_names(cluster_summary)
df['Cluster Name'] = df['Cluster'].map(cluster_names)

# Summary of Clusters
for cluster_id, cluster_name in cluster_names.items():
    print(f"Cluster {cluster_id}  {cluster_name}:")
    cluster_data = cluster_summary[cluster_summary['Cluster'] == cluster_id]
    print(cluster_data.to_string(index=False))
    print("\n")

# Display DataFrame with Cluster Names
print(df.head())
```
